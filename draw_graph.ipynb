```python

def draw_cumulative_score(filename,title):
    df = pandas.read_csv(filename, header=None)
    data = df.values
    # x axis values 
    x =  [float(x[0]) for x in data]
    #print (x)
    # corresponding y axis values 
    y = [float(x[1]) for x in data]
    #print (y)
    series = Series(y)
    print ('Min', series.min(), 'Max', series.max(), 'mean', series.mean())
    # plotting the points  
    plt.plot(x, y) 

    # naming the x axis 
    plt.xlabel('Episodes') 
    # naming the y axis 
    plt.ylabel('Cumulative Score per episode') 

    # giving a title to my graph 
    plt.title(title) 

    # show a legend on the plot 
    # plt.legend() 

    # function to show the plot 
    plt.show() 
draw_cumulative_score('dqn_score_card.csv', 'DQN Agent')
```

```python

def draw_cumulative_score_4(filename,title):
    df = pandas.read_csv(filename, header=None)
    data = df.values[400:600]
    # x axis values 
    x =  [float(x[0]) for x in data]
    y = [float(x[1]) for x in data]
    # Tail-rolling average transform
    series = Series(y)
    rolling = series.rolling(window=3)
    rolling_mean = rolling.mean()
    print ('Min', series.min(), 'Max', series.max(), 'mean', series.mean())
    #print(rolling_mean.head(10))
    # plot original and transformed dataset
    #series.plot()
    plt.plot(x, rolling_mean, color='red') 
    plt.title(title) 
     # naming the x axis 
    plt.xlabel('Episodes') 
    # naming the y axis 
    plt.ylabel('Cumulative Score per timestep') 
    #rolling_mean.plot(color='red')
    plt.show()
```

```python
def draw_cumulative_score_5(filename1, filename2, label1, label2, title):
    plt.figure(figsize=(10,5))
    
    df1 = pandas.read_csv(filename1, header=None)
    data1 = df1.values
    df2 = pandas.read_csv(filename2, header=None)
    data2 = df2.values
    # x axis values 
    x =  [float(x[0]) for x in data1]
    #print (x)
    # corresponding y axis values 
    y1 = [float(x[1]) for x in data1]
    y2 = [float(x[1]) for x in data2]
  
    plt.plot(x, y1, label = label1) 
    plt.plot(x, y2, label = label2) 
    y3 = [np.mean(y1) for x in data1]
    plt.plot(x, y3, label = 'Mean score - ' + label1) 
    y4 = [np.mean(y2) for x in data1]
    plt.plot(x, y4, label = 'Mean score - ' + label2) 
    plt.ylabel('Cumulative Score per episode') 
    
    # naming the x axis 
    plt.xlabel('Episodes') 

    # giving a title to my graph 
    plt.title(title) 

    # show a legend on the plot 
    plt.legend() 

    # function to show the plot 
    plt.show() 
draw_cumulative_score_5('dqn_play_score_card_100.csv', 'random_score_card_100.csv', 'DQN Agent', 'Random Agent', 'DeepQ Agent vs Random')
```

```python
def draw_cumulative_score_6(filenames, labels):
    plt.figure(figsize=(16,8))
    
    for idx, filename in enumerate(filenames):
        df = pandas.read_csv(filename, header=None)
        data = df.values
        x =  [float(x[0]) for x in data]
        y = [float(x[1]) for x in data]
        label = labels[idx]
        y_mean = [np.mean(y) for x in data]
        plt.plot(x, y, label = labels[idx]) 
        plt.plot(x, y_mean, label = 'Mean score - ' + label)
        

    # naming the x axis 
    plt.xlabel('Episodes') 

    # show a legend on the plot 
    plt.legend() 

    # function to show the plot 
    plt.show() 
```
